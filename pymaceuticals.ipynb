## Observations and Insights 

```python
#%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)

# exam common column names so join can be performed later
mouse_metadata_df.head()
study_results_df.head()
# After checking the columns for both tables, mouse ID can be the primary key to perform the join

# Combine the data into a single dataset
mater_table = pd.merge(mouse_metadata_df,study_results_df, on= "Mouse ID")
# Display the data table for preview
mater_table.sample(10)
```

```python
# Checking the number of mice.
len(mater_table["Mouse ID"])
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

naughty_mouse = mater_table.groupby(["Mouse ID","Timepoint"]).size().reset_index(name = "Frequency")
naughty_mouse = naughty_mouse.sort_values(["Frequency"],ascending=[False])
naughty_mouse.head(20)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# above table shows that g989 repeat 5 times 
naughty_mouse_repeat = naughty_mouse.loc[naughty_mouse["Frequency"] != 1]
naughty_mouse_repeat.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
unique_master_table = mater_table.drop_duplicates(["Mouse ID","Timepoint"],keep= 'last')
```

```python
# Checking the number of mice in the clean DataFrame.
len(unique_master_table)
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
time_drug = unique_master_table.groupby(["Drug Regimen", "Timepoint"])
Capo_mean_tumor = time_drug.mean()
Capo_mean_tumor.head(20)
```

```python



timepoints = np.arange(0, 50, 5)
tumor = Capo_mean_tumor["Tumor Volume (mm3)"]
drug = tumor.loc["Capomulin", :]
capo_tumor_plot = plt.plot(timepoints, drug,label = "Capomulin")


plt.title("Average Tumor Volume Over Treatment")
plt.xlabel("Days")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best")

plt.show()
```

```python
Capo_mean_tumor.head(10)
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
Capo_selected = Capo_mean_tumor.head(10)
weight = Capo_selected["Weight (g)"]
tumor_volume = Capo_selected["Tumor Volume (mm3)"]
plt.scatter(weight, tumor_volume, marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.title("Weight vs. average tumor volume")
plt.xlabel("Weight")
plt.ylabel("Tumor Volume (mm3)")


plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
wt_slope, wt_int, wt_r, wt_p, wt_std_err = st.linregress(weight, tumor_volume)
wt_fit = wt_slope * weight + wt_int

plt.scatter(weight, tumor_volume, marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.plot(weight,wt_fit,"--")
plt.title("Weight vs. average tumor volume")
plt.xlabel("Weight")
plt.ylabel("Tumor Volume (mm3)")


plt.show()
```
